```python
### Calculation of gene level profiles from the Hela hplm guide level profiles for the PERISCOPE manuscript. ###
### Script by Meraj Ramezani(mramezan@broadinstitute.org) ###
# Import relevant librariesimport pandas as pdimport pandas as pd
import pandas as pd
import numpy as np
from pycytominer import normalize, feature_select ,aggregate,consensus
from pycytominer.cyto_utils import output,infer_cp_features
from pycytominer.cyto_utils.util import (
    get_pairwise_correlation,
    check_correlation_method,
    infer_cp_features,
)
```

```python
# load normalized_feature_selected profiles for the Hela hplm WGS
filename = '(input the file address)/20210422_6W_CP257_guide_normalized_feature_select_median_merged_ALLBATCHES___HPLM___ALLWELLS.csv.gz'
hplm_median_df = pd.read_csv(filename)
hplm_median_df
```

```python
# Filter out irrelavant guides
filename = '(input the file address)/wgs_guide_list.csv'
guide_df = pd.read_csv(filename)
hplm_guide_df = guide_df.query('dialout == 1 | dialout ==3')
hplm_guide_list = list(hplm_guide_df['sgRNA'])
hplm_median_df = hplm_median_df[hplm_median_df["Metadata_Foci_Barcode_MatchedTo_Barcode"].isin(hplm_guide_list)]
hplm_median_df
```

```python
# Perform profile aggregation
aggregate_columns = ['Metadata_Foci_Barcode_MatchedTo_GeneCode']
aggregate_features = 'infer'
aggregate_operation = 'median'

hplm_median_gene_df = aggregate(
            population_df=hplm_median_df,
            strata=aggregate_columns,
            features=aggregate_features,
            operation=aggregate_operation,
            )
hplm_median_gene_df
```

```python
# Save the gene level profiles
hplm_median_gene_df.to_csv('20210422_6W_CP257_guide_normalized_feature_select_median_merged_ALLBATCHES___HPLM___ALLWELLS_gene_aggregated.csv.gz',index=False)
```
